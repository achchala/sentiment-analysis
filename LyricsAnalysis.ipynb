# Natural Language Processing - Lyric Analysis and Visualization

### Cleaning and Preparing Data

```python
import numpy as np 
import pandas as pd 
import re
import os
```

```python
path_to_album = './data/Albums/'
album_list = [os.path.join(path_to_album, album) for album in os.listdir(path_to_album)]
album_list
```

The tabular folder has tables for each album and within each album folder we have songs. So, now, we will use the tabular table to map the paths above.



```python
tabular_path = './data/Tabular/'
songs_list = [os.path.join(tabular_path, song) for song in os.listdir(tabular_path)]
songs_list
```

```python
album_df = pd.read_csv('./data/Albums.csv', index_col=0)
album_df.head()
```

```python
album_df['Track_path'] = album_df.Albums.apply(lambda x : os.path.join(tabular_path, re.sub('[^a-zA-Z0-9_]','_',''.join(x.split()))+'.csv'))
album_df.head()
```

```python
album_df.loc[0,'Track_path']
```

Fetching all songs within each album

```python
df = pd.DataFrame()
for ind, album in enumerate(album_df['Track_path']):
    try:
        track_df = pd.read_csv(album)
        track_df['Album_ID'] = album_df.loc[ind]['ID']
        track_df['Album'] = album_df.loc[ind]['Albums']
        track_df['Album_Path'] = os.path.join(path_to_album, re.sub('[^a-zA-Z0-9_]','_',''.join(album_df.loc[ind]['Albums'].split())))
        df = pd.concat([df, track_df])
    except Exception as e:
        print('No songs in album: ',album )
        pass

df
```

```python
df.reset_index(drop=True, inplace=True)
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.loc[30,'Album_Path']
```

```python
def get_lyrics(x):
    path = (x['Album_Path'] + '/' +re.sub('[^a-zA-Z0-9_]','_',''.join(x['Tracks'].split())) + '.txt').replace('_s', 's').replace('the_ladieslunching_chapter', 'theladieslunchingchapter').replace('Carolina_FromTheMotionPicture_WhereTheCrawdadsSing__','Carolina_FromTheMotionPictureWhereTheCrawdadsSing_').replace('evermore_the_foreveristhesweetestcon_chapter','evermore_theforeveristhesweetestconchapter').replace('willow_90strendremix_.txt','/willow_90_strendremix_.txt')
    try:
        with open(path, 'r') as f:        
            txt = f.read()
            f.close()
        return txt
    except Exception as e:
        print(path)
        pass
```

```python
df['lyrics'] = df.apply(get_lyrics, axis=1)
df.head()
```
