Fetching all songs within each album

```python
df = pd.DataFrame()
for ind, album in enumerate(album_df['Track_path']):
    try:
        track_df = pd.read_csv(album)
        track_df['Album_ID'] = album_df.loc[ind]['ID']
        track_df['Album'] = album_df.loc[ind]['Albums']
        track_df['Album_Path'] = os.path.join(path_to_album, re.sub('[^a-zA-Z0-9_]','_',''.join(album_df.loc[ind]['Albums'].split())))
        df = pd.concat([df, track_df])
    except Exception as e:
        print('No songs in album: ',album )
        pass

df
```

```python
def get_lyrics(x):
    path = (x['Album_Path'] + '/' +re.sub('[^a-zA-Z0-9_]','_',''.join(x['Tracks'].split())) + '.txt').replace('_s', 's').replace('the_ladieslunching_chapter', 'theladieslunchingchapter').replace('Carolina_FromTheMotionPicture_WhereTheCrawdadsSing__','Carolina_FromTheMotionPictureWhereTheCrawdadsSing_').replace('evermore_the_foreveristhesweetestcon_chapter','evermore_theforeveristhesweetestconchapter').replace('willow_90strendremix_.txt','/willow_90_strendremix_.txt')
    try:
        with open(path, 'r') as f:        
            txt = f.read()
            f.close()
        return txt
    except Exception as e:
        print(path)
        pass
```

```python
df['lyrics'] = df.apply(get_lyrics, axis=1)
df.head()
```

Cleaning Lyrics

```python
df.loc[0,'lyrics']
```

```python
def remove_lyrics_header(lyrics):
    if lyrics is None:
        return None  # Return None if lyrics is None
    lyrics_list = lyrics.split('Lyrics', 1)  # split the string where 'Lyrics' occurs (because random info before)
    if len(lyrics_list) > 1:
        return lyrics_list[1]  # and, return the substring after 'Lyrics'
    else:
        return lyrics  # return the original string (whole string) if 'Lyrics' is not found


df["lyrics"] = df.lyrics.apply(lambda x: remove_lyrics_header(x))
```

```python
df.head()
```

Adding a field of the labels for parts of the song, to get a better idea of the songs' structures.

```python
import re

def extract_labels(text):
    if text is None:
        return None  # Return None if text is None
    labels = re.findall(r'\[(.*?)\]', text)
    return labels

df["labels"] = df.lyrics.apply(lambda x: extract_labels(x))
```

```python
df
```

```python
df.loc[0,'labels']
```

```python
def clean_labels(labels):
    if labels is None:
        return None  # Return None if labels is None
    cleaned_labels = []
    for label in labels:
        if len(label.split(':')) > 1:
            cleaned_label = re.sub(r':.*$', '', label).strip()
            cleaned_labels.append(cleaned_label)
        else:
            cleaned_labels.append(label)
    return cleaned_labels
```

```python
df["labels"] = df.labels.apply(lambda x: clean_labels(x))
```
